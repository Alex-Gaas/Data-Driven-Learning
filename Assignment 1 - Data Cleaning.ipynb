```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('Assignment 1 - steps data.csv', sep = ';')
df1.head(5)
```

```python
df2 = pd.read_csv('Assignment 1 - survey data.csv')
df2.head(5)
```

```python
df = pd.merge(df2, df1, on = 'id')
df.head(5)
```

```python
df['waist'].value_counts()
```

```python
df['waist'] = df['waist'].apply(pd.to_numeric, errors='coerce')
def between_45130(x):
    if(x < 45.0 or x > 130.0): 
        return float('NaN')
    else: 
        return x
df['waist'] = df['waist'].apply(between_45130)
```

```python
sns.distplot(df['waist'].dropna(), kde=False)
plt.title('Waist size of Fitbit users')
plt.xlabel('Waist size (in cm)')
plt.show()
```
