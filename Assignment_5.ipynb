```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr
```

```python
df = pd.read_csv('funda-amsterdam.csv')
df.head()
```

```python
df_subset = df[['rooms', 'bedrooms', 'price']]
df_subset.head()
```

```python
sns.pairplot(df_subset)
```

```python
df_subset["price"].min()
```

```python
df_subset2 = df_subset.drop(df_subset[df_subset.price < 30000].index)
df_subset2["price"].min()
```

```python
sns.pairplot(df_subset2)
```

```python
sns.scatterplot(data=df_subset2, x="price", y="rooms")
```

It is a linear correlation: when the amount of rooms are higher, the price is higher. When the amount of rooms is lower, the price is lower.

```python
pearsonr(df_subset2['price'],df_subset2['rooms'])
```

Pearson's r is between 0.3 and 0.6 (0.54) so it is _moderate_
